```python
import numpy as np
import healpy as hp
import pandas as pd
from scipy import interpolate
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter, NullFormatter

from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.io import fits
from astropy.table import Table

from tomographer.runtime_script import tomo_utils as utils
import yaw
```

```python
from astropy.cosmology import FlatLambdaCDM

h      = 0.6737
rp_min = 0.5/h # Mpc
rp_max = 10./h # Mpc
max_theta_max = 15. # deg
cosmo = FlatLambdaCDM(H0=h*100., Om0=0.3132, Tcmb0=2.725) # Planck18

z_edges, z_ctrs = utils.z_binning_log1pz(0., 4.2, 160)
delta_z = z_edges[1:] - z_edges[:-1]
print('dense z bin centers [160 log(1+z) bins in 0<z<4.2]:\n', z_ctrs, '\n')

rebinned_z_edges, rebinned_z_ctrs = utils.z_binning_log1pz(0., 4.2, 40)
print('rebinned z bin centers [40 log(1+z) bins in 0<z<4.2]:\n', rebinned_z_ctrs, '\n')
```

```python
zmin = np.array([0.1, 0.75, 1.4])
zmax = np.array([0.3, 1.0, 1.8])
```

# YAW

```python
patch_num = 32
cache_dir = "source_catalog/yaw_cache"  # create as needed

config = yaw.Configuration.create(
    rmin=500.0/h,  # can also be a list of lower scale limits
    rmax=10000.0/h, # can also be a list of upper scale limits
    # unit="kpc"  # defaults to angular diameter distance, but angles and
                  # comoving transverse distance are supported
    rweight=-1.8,     # if you want to weight pairs by scales
    resolution=50,  # resolution of weights in no. of log-scale bins
    method="logspace",
    cosmology=cosmo,
    edges=rebinned_z_edges,  # provide your custom bin edges
)

for i_z in range(3):
    zlow = zmin[i_z]
    zup = zmax[i_z]
    reference_path = 'source_catalog/' f'test_data_{zlow}_{zup}.pqt'
    ref_rand_path = 'source_catalog/' f'test_rand_{zlow}_{zup}.pqt'

    cat_ref_rand = yaw.Catalog.from_file(
        cache_directory=f"{cache_dir}/ref_rand",
        path=ref_rand_path,
        ra_name="RA",
        dec_name="DEC",
        redshift_name="Z",  # required for reference
        patch_num=patch_num,
        overwrite=True,
        progress=True,  # shows a progress bar, default: False
    )
    
    # extract the patch centers to use these for all following catalogs
    patch_centers = cat_ref_rand.get_centers()
    
    cat_reference = yaw.Catalog.from_file(
        cache_directory=f"{cache_dir}/reference",
        path=reference_path,
        ra_name="RA",
        dec_name="DEC",
        redshift_name="Z",  # required for reference
        patch_centers=patch_centers,  # use previously computed centers
        overwrite=True,
        progress=True,  # shows a progress bar, default: False
    )

    cts_ss_list = yaw.autocorrelate(
        config,
        cat_reference,
        cat_ref_rand,
        count_rr=False,
        progress=True,  # shows a progress bar, default: False
    )

    cts_ss = cts_ss_list[0]
    cts_ss.to_file(f"w_ss_{zlow}_{zup}.hdf5")
```

# Validation

```python
gamma = -0.8

def sqrt_error(value: float, error: float):
    res_val = np.sqrt(value)
    res_err = 0.5 * (error / value) * res_val
    return res_val, res_err
```

```python
#w_ss_ref_40bins_hacked.hdf5
def b_intp_yaw(filename):
    
    cts_ss_40bins = yaw.CorrFunc.from_file(f"{filename}")
    yaw_w_rr_40bins = cts_ss_40bins.sample().data
    yaw_w_rr_40bins_err = cts_ss_40bins.sample().error
    
    deg_per_Mpc = cosmo.arcsec_per_kpc_proper(rebinned_z_ctrs).to(u.deg/u.Mpc).value
    theta_min = deg_per_Mpc * rp_min
    theta_max = deg_per_Mpc * rp_max
    theta_max[theta_max>max_theta_max] = max_theta_max
    rmax, rmin = np.radians(theta_max), np.radians(theta_min)
    X = (rmax**(gamma+1) - rmin**(gamma+1)) / (gamma+1)
    yaw_w_rr_40bins *= X
    yaw_w_rr_40bins_err *= X
    
    matter_filepath = ('/Users/yuvoon/Documents/clustering_redshift/dev_v08/tomographer/precalculated_data/cosmology/planck_2018/wmbar_prp_0.5-10Mpch_gamma-0.8_15.0degtmax_40log1pzbins_multi_beam_filtering.fits')
    wm = Table.read(matter_filepath)
    
    # ==== Beam and Filtering Info for w_m Correction ====
    beam_tag = 'NS2048' # The minimum would be the pixel window
    # w_m_beam_fwhm_arcmin_lst = np.array([2., 3., 4., 4.3, 5., 5.5, 6., 7., 7.1, 8., 9., 10., 12., 15., 20., 25., 30.]) # implemented in w_m
    filtering_tag = f'inf_deg_filtered'
    
    # ==== Matter Clustering Correcttion Preparation ====
    # Ref-ref auto is effectively pixalized reference map cross reference catalog (pixel-window x point sources)
    f_wm_ref_ref = interpolate.interp1d(wm['z'], wm['wmbar_NS2048']) # For ref-ref auto and b_ref calculation
    wm_40bins = list(f_wm_ref_ref(rebinned_z_ctrs))
    
    yaw_40bins = (yaw_w_rr_40bins/wm_40bins * np.diff(rebinned_z_edges))
    yaw_40bins_err = yaw_w_rr_40bins_err/wm_40bins * np.diff(rebinned_z_edges)
    yaw_40bins, yaw_40bins_err = sqrt_error(yaw_40bins, yaw_40bins_err)

    bt_intp = interpolate.interp1d(rebinned_z_ctrs, yaw_40bins, fill_value="extrapolate")
    bt_intp = list(bt_intp(z_ctrs))
    bt_intp = np.array(bt_intp)
    bt_intp[bt_intp<0] = 0
    yaw_40bins = bt_intp

    return bt_intp, yaw_40bins
```

```python
key_col = 'dNdz_b'
styles = {160: {'ms': 3.5, 'capsize': 2}, 40: {'ms': 6, 'capsize': 3.5}}
area = [10522.83, 7302.39, 5126.97]
filenames = ['source_catalog/test_data_0.1_0.3', 'source_catalog/test_data_0.75_1.0', 'source_catalog/test_data_1.4_1.8']

fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 6), gridspec_kw={'height_ratios': [1,2,1]}, dpi=250)
ax = axs.ravel()
plt.subplots_adjust(hspace=0.)

for i in range(len(filenames)):

    fn =  filenames[i]
    file = fits.open(f'{fn}.fits')
    w = file[1].data

    zlow, zup = re.findall(pattern, fn)
    zlow = float(zlow)
    zup = float(zup)
    
    test_dat = pd.read_parquet(f'test_data_{zlow}_{zup}.pqt')
    test_num = len(test_dat)

    y1 = w[key_col] 
    y1err = w['{}_err'.format(key_col)] 
    # ax[1].errorbar(w['z']+1, y1, yerr=y1err, fmt='.', lw=0.8, zorder=i+1, **styles.get(len(y1)), c=f'C{i}')   
    ax[1].fill_between(w['z']+1, y1-y1err, y1+y1err, color=f'C{i}', alpha=.5)    
    ax[1].axhline(0, color='grey', ls='--', zorder=-10)

    b_160bins, b_40bins = b_intp_yaw(f"w_ss_{zlow}_{zup}.hdf5")
    
    if len(y1)==40:
        bins = rebinned_z_edges
        b_yaw = b_40bins
    else:
        bins = z_edges
        b_yaw = b_160bins
    
    counts, edges = np.histogram(test_dat['Z'], bins=bins)
    sel = np.where(b_yaw>0)
    counts = counts/file[0].header['TOTAL_AREA']
    truth = counts/(w['z_up']-w['z_low'])
    truth[sel] *= b_yaw[sel]
    
    ax[0].plot(w['z']+1, (y1-truth)/y1err, c=f'C{i}', lw=3, alpha=.5)
    ax[1].plot(w['z']+1, truth, lw=2, zorder=i, color='black')
    
    ax[2].plot(z_ctrs+1, b_160bins, marker='o', ms=4, c=f'C{i}', mfc='white')
    ax[2].plot(rebinned_z_ctrs+1, b_40bins, 'o-', ms=4, lw=2, zorder=0, c=f'C{i}', markeredgecolor='black')

    for k in range(3):
        ax[k].set_xscale('log')
        ax[k].xaxis.set_major_formatter(StrMethodFormatter('{x:.0f}'))
        ax[k].xaxis.set_minor_formatter(NullFormatter())
        # zticks = np.array([0, 0.5, 1,1.5,  2, 2.5, 3, 3.5, 4.])
        # zticks_str = ['0', '0.5', '1', '', '2', '', '3', '', '4']
        zticks = np.round(np.hstack((np.arange(0, 1.1, 0.1), np.arange(1.2, 4.1, 0.2))),1)
        zticks_str = [
            '0.5' if val == 0.5 else (str(int(val)) if val.is_integer() else '') 
            for val in zticks
        ]
        
        ax[k].set_xticks(1.+zticks)
        ax[k].set_xlim([.97, 5.3])
            
        ax[k].set_xticklabels(zticks_str)

    ax[0].set_ylim(-5.6,5.6)
    ax[0].axhline(0, color='grey', ls='--', zorder=-1)
    
    ylabel = r'b\ \mathrm{d}N/\mathrm{d}z\ \mathrm{[\#/deg^2]}'
    ax[1].set_ylabel(r'${}$'.format(ylabel))
    ax[2].set_ylabel(r'$b_t$')
    ax[2].set_xlabel(r'$z$', y=0.05)
    # plt.yscale('log')
    # ax[1].set_ylim(-1000,5000)

plt.show()
```
