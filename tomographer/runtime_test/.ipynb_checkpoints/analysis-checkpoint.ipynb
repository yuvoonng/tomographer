```python
import yaw
cts_ss_160bins = yaw.CorrFunc.from_file("w_ss_ref_160bins_nosel.hdf5")
cts_ss_40bins = yaw.CorrFunc.from_file("w_ss_ref_40bins_hacked.hdf5")
cts_ss_20bins = yaw.CorrFunc.from_file("w_ss_ref_20bins_nosel.hdf5")
sdss_z0p15_40bins_nosel = fits.open('sdss_z0p15_40bins_nosel_newcorr.fits')[1].data
sdss_z0p15_160bins_nosel = fits.open('sdss_z0p15_160bins_nosel.fits')[1].data

def spatcorr_160bins_func(x):
    # a,b,c,d = 0.43, 1.04, 0.05, 1.01
    a,b,c,d = 0.49, 1.05, 0.06, 1.01
    return d + (d-a)/(1+np.exp(-(x-b)/c))
    
zless = np.argmin(np.abs(z_ctrs-3))
zmore = np.argmin(np.abs(z_ctrs-.05))
spatcorr_160bins = (np.array(spatcorr_160bins_func(z_ctrs[zmore:zless+1])))

yaw_w_rr_160bins = cts_ss_160bins.sample().data
yaw_w_rr_160bins_err = cts_ss_160bins.sample().error
# sdss_z0p15_160bins_nosel['b_r_final'][zmore:zless] /= np.sqrt(spatcorr_160bins)

def spatcorr_40bins(x):
    a, b, c, h, k = 2.80, 2.57, 0.19, -0.29, 0.88
    # a, b, c, h, k = 0.44, 2.28, 0.13, 0.71 , 0.88
    return a * c**(b*x-h) + k

zless = np.argmin(np.abs(rebinned_z_ctrs-3))
zmore = np.argmin(np.abs(rebinned_z_ctrs-.05))
spatcorr_40bins = (np.array(spatcorr_40bins(rebinned_z_ctrs[zmore:zless+1])))

yaw_w_rr_40bins = cts_ss_40bins.sample().data
yaw_w_rr_40bins_err = cts_ss_40bins.sample().error
# sdss_z0p15_40bins_nosel['b_r_final'][zmore:zless] /= np.sqrt(spatcorr_40bins)

yaw_w_rr_20bins = cts_ss_20bins.sample().data
```

```python
gamma = -0.8
deg_per_Mpc = cosmo.arcsec_per_kpc_proper(z_ctrs).to(u.deg/u.Mpc).value
theta_min = deg_per_Mpc * rp_min
theta_max = deg_per_Mpc * rp_max
theta_max[theta_max>max_theta_max] = max_theta_max
rmax, rmin = np.radians(theta_max), np.radians(theta_min)
X = (rmax**(gamma+1) - rmin**(gamma+1)) / (gamma+1)

yaw_w_rr_160bins *= X
yaw_w_rr_160bins_err *= X

deg_per_Mpc = cosmo.arcsec_per_kpc_proper(rebinned_z_ctrs).to(u.deg/u.Mpc).value
theta_min = deg_per_Mpc * rp_min
theta_max = deg_per_Mpc * rp_max
theta_max[theta_max>max_theta_max] = max_theta_max
rmax, rmin = np.radians(theta_max), np.radians(theta_min)
X = (rmax**(gamma+1) - rmin**(gamma+1)) / (gamma+1)
yaw_w_rr_40bins *= X
yaw_w_rr_40bins_err *= X

deg_per_Mpc = cosmo.arcsec_per_kpc_proper(z_ctrs_20bins).to(u.deg/u.Mpc).value
theta_min = deg_per_Mpc * rp_min
theta_max = deg_per_Mpc * rp_max
theta_max[theta_max>max_theta_max] = max_theta_max
rmax, rmin = np.radians(theta_max), np.radians(theta_min)
X = (rmax**(gamma+1) - rmin**(gamma+1)) / (gamma+1)
yaw_w_rr_20bins *= X
```

```python
def sqrt_error(value: float, error: float):
    res_val = np.sqrt(value)
    res_err = 0.5 * (error / value) * res_val
    return res_val, res_err

yaw_160bins = (yaw_w_rr_160bins/wm_160bins * np.diff(z_edges))
yaw_160bins_err = yaw_w_rr_160bins_err/wm_160bins * np.diff(z_edges)
yaw_160bins, yaw_160bins_err = sqrt_error(yaw_160bins, yaw_160bins_err)

yaw_40bins = (yaw_w_rr_40bins/wm_40bins * np.diff(rebinned_z_edges))
yaw_40bins_err = yaw_w_rr_40bins_err/wm_40bins * np.diff(rebinned_z_edges)
yaw_40bins, yaw_40bins_err = sqrt_error(yaw_40bins, yaw_40bins_err)

sdss_160bins = np.sqrt(sdss_z0p15_160bins_nosel['w_rr']/wm_160bins * np.diff(z_edges))
sdss_40bins = np.sqrt(sdss_z0p15_40bins_nosel['w_rr']/wm_40bins * np.diff(rebinned_z_edges))
```

```python
def quotient_err(x, dx, y, dy):

    q = x / y
    rel_x = dx / x
    rel_y = dy / y
    rel_q = np.sqrt(rel_x**2 + rel_y**2)
    dq = np.abs(q * rel_q)
    
    return q, dq
    
from scipy.optimize import curve_fit

def exp_func(x, a, b, c, h, k):
    return a * c**(b*x-h) + k
    
zless = np.argmin(np.abs(rebinned_z_ctrs-3))
zmore = np.argmin(np.abs(rebinned_z_ctrs-.05))
x_data_40bins = rebinned_z_ctrs[zmore:zless+1]
x,dx,y,dy = sdss_z0p15_40bins_nosel['w_rr'], sdss_z0p15_40bins_nosel['w_rr_err'], yaw_w_rr_40bins, yaw_w_rr_40bins_err
q, dq = quotient_err(x,dx,y,dy)
y_data = q[zmore:zless+1]
y_err = dq[zmore:zless+1]

popt_40bins, _ = curve_fit(exp_func, x_data_40bins, y_data, sigma=y_err, absolute_sigma=True)
# popt

def sigmoid_func(x, a , b , c, d):
    return d + (d-a)/(1+np.exp(-(x-b)/c))

zless = np.argmin(np.abs(z_ctrs-3))
zmore = np.argmin(np.abs(z_ctrs-.05))
x_data_160bins = z_ctrs[zmore:zless+1]
x,dx,y,dy = sdss_z0p15_160bins_nosel['w_rr'], sdss_z0p15_160bins_nosel['w_rr_err'], yaw_w_rr_160bins, yaw_w_rr_160bins_err
q, dq = quotient_err(x,dx,y,dy)
y_data = q[zmore:zless+1]
y_err = dq[zmore:zless+1]

popt_160bins, _ = curve_fit(sigmoid_func, x_data_160bins, y_data, sigma=y_err, absolute_sigma=True)

print(popt_40bins, popt_160bins)
```
