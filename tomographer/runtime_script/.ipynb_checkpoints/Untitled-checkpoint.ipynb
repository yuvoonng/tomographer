```python
from pathlib import Path
import tomllib
import tomli_w

CONFIG_PATH = Path.home()  / ".precal_data_path.toml"

def detect_data_path():
    # auto-detect location
    candidates = [
        Path("./Precalculated_Data"),
        Path("../Precalculated_Data"),
    ]

    for c in candidates:
        if c.is_dir():
            return c.resolve()

    raise ValueError("Precalculated_Data not found.")
```

```python
def init_config():
    data_path = detect_data_path()

    config = {
        "data": {
            "precalculated_data_path": str(data_path)
        }
    }

    with open(CONFIG_PATH, "wb") as f:
        tomli_w.dump(config, f)

    print(f"[INIT] Config saved to {CONFIG_PATH}")


def load_or_init_config(user_config=None):
    # 1. user explicitly passed config
    if user_config:
        with open(user_config, "rb") as f:
            return tomllib.load(f)

    # 2. first run
    if not CONFIG_PATH.exists():
        init_config()

    # 3. load stored config
    with open(CONFIG_PATH, "rb") as f:
        return tomllib.load(f)
```

```python
load_or_init_config()
```
